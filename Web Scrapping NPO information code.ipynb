# Pagination

```python
base_url
```

```python
pages = []
for i in range(2,51):
    page = base_url + 'sort:review_count/direction:desc/page:'+str(i)
    pages.append(page)
pages
```

```python
for page in pages:
    print(page)
    page_response = requests.get(page)
    page_html = page_response.content
    page_soup = BeautifulSoup(page_html,'lxml')
    page_body = page_soup.find_all('body')
    for i in range(len(page_body)):
        page_links = page_body[i].find_all('h2',{'property':'name'})
        for i in range(len(page_links)):
            org = page_links[i].find('a').get('href')
            orgs.append(org)
            print(org)
orgs
```

```python
#join the url
full_urls = [urljoin(base_url,org) for org in orgs]
full_urls
```

## Extract information

```python
orgs = []
cats = []
emails = []
phones = []
addresses = []
cities = []
i = 1
for url in full_urls:
    print (url)
    print (i)
    response = requests.get(url)
    html = response.content
    soup = BeautifulSoup(html,'lxml')
    body = soup.find('body')
    info = body.find('div',{'id':'gnp-np-profile'})
    info2 = info.find('div',{'class':'col-xs-16 col-sm-6 col-md-4'})
    info3 = info2.find('div',{'id':'np-info'})
    
    header = info.find('div',{'class':'gnp-np-profile-head'})
    header2 = header.find('h1').text
    orgs.append(header2)
    print(header2)
    
    cat = info.find('div',{'class':'np-overview gnp-show-more'})

    try:
        cat2 = cat.find_all('a')
        temp_cat = []
        for item in cat2:
            temp_cat.append(item.text)
        cat4 = ','.join(temp_cat)
    except:
        cat4 = 'None'
    cats.append(cat4)
    print(cat4)
    
    try:
        email = info3.find('a',{'itemprop':'email'}).get('href')
    except:
        email = 'None'
    
    try:
        phone = info3.find('span',{'class':'telephone'}).text
    except:
        phone = 'None'
   
    try:
        address = info3.find('span',{'itemprop':'streetAddress'}).text
    except:
        address = 'NOne'
    
    try:
        city = info3.find('span',{'itemprop':'addressLocality'}).text
    except:
        city = 'None'
    print(email,phone,address,city)
    
    emails.append(email)
    phones.append(phone)
    addresses.append(address)
    cities.append(city)
    i = i + 1
```

```python
import pandas as pd
```

```python
np_info = pd.DataFrame()

np_info['Organization'] = orgs
np_info['Categories'] = cats
np_info['Email'] = emails
np_info['Phone'] = phones
np_info['Address'] = addresses
np_info['City'] = cities

np_info
```

```python
np_info_copy = np_info
np_info_copy
```

```python
np_info_copy['Email'] = np_info_copy['Email'].replace('mailto:','',regex=True)
np_info_copy
```
